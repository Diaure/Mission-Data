2. **Calcul des votes des candidats fictifs `blancs et nuls`, `blancs, nuls et abstentions` (chaque dataset)**
* Exp = votes exprimés
* % Ins = pourcentage des votes par rapport aux inscrits
* % Exp = pourcentage des votes par rapport aux exprimés
* % Vot = pourcentage des votes par rapport aux votants

```python
# 2022
elections_2022['Blancs et nuls'] = elections_2022['Blancs'] + elections_2022['Nuls']
elections_2022['BlNuls/Exp'] = round((elections_2022['Blancs et nuls'] / elections_2022['Exprimés']) * 100, 2)

elections_2022['% BlNuls/Ins'] = round((elections_2022['Blancs et nuls'] / elections_2022['Inscrits']) * 100, 2)
elections_2022['% BlNuls/Vot'] = round((elections_2022['Blancs et nuls'] / elections_2022['Votants']) * 100, 2)

elections_2022['Blancs_nuls et abstentions'] = elections_2022['Blancs et nuls'] + elections_2022['Abstentions']
elections_2022['BlNulA/Exp'] = round((elections_2022['Blancs_nuls et abstentions'] / elections_2022['Exprimés']) * 100, 2)

elections_2022['% BlNulA/Ins'] = round((elections_2022['Blancs_nuls et abstentions'] / elections_2022['Inscrits']) * 100, 2)
elections_2022['% BlNulA/Vot'] = round((elections_2022['Blancs_nuls et abstentions'] / elections_2022['Votants']) * 100, 2)


```

```python
# calculer les scores moyens par départements afin d'avoir une ligne par département
tour1_22_moyenne = firsttour_2022_elections.groupby('Libellé du département')[['Blancs et nuls', 'Blancs_nuls et abstentions', 'Nathalie ARTHAUD', 'Fabien ROUSSEL', 'Emmanuel MACRON', 'Jean LASSALLE', 
                                                         'Marine LE PEN', 'Éric ZEMMOUR', 'Jean-Luc MÉLENCHON', 'Anne HIDALGO', 'Yannick JADOT', 'Valérie PÉCRESSE', 'Philippe POUTOU', 
                                                         'Nicolas DUPONT-AIGNAN']].mean().reset_index()
tour1_22_moyenne = tour1_22_moyenne.rename({"Libellé du département": "Départements"}, axis=1)
tour1_22_moyenne
```

```python
import numpy as np
tour1_22_moyenne_clean = tour1_22_moyenne[
    ~np.isinf(tour1_22_moyenne['Blancs et nuls']) & 
    ~np.isinf(tour1_22_moyenne['Blancs_nuls et abstentions'])
]
tour1_22_moyenne_clean
```

```python
import requests
def geocode(adresse):
    url = f"https://api-adresse.data.gouv.fr/search/?q={adresse}"
    response = requests.get(url)
    data = response.json()
    if 'features' in data and len(data['features']) > 0:
     coords = data['features'][0]['geometry']['coordinates']
     return coords[::-1]
    else:
     return None
```

```python
# rechercher les coordonnées pour chaque département pour l'affichage dans la carte géographique

meilleurs_candidats_tour1_22 = meilleurs_candidats_tour1_22.reset_index()
meilleurs_candidats_tour1_22['coordonnées'] = meilleurs_candidats_tour1_22['Départements'].apply(geocode)
```

```python
meilleurs_candidats_tour1_22["longitude"]= meilleurs_candidats_tour1_22["coordonnées"].str[1]
meilleurs_candidats_tour1_22["latitude"]= meilleurs_candidats_tour1_22["coordonnées"].str[0]
```
